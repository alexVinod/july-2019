```python
# Visualisation
# Linear graph
import matplotlib.pyplot as plt
plt.title("Basic Linear Graph")
plt.plot([1,2,3],[1,2,3])
plt.show()
```

```python
# Graph to plot distance and time
distance = list(range(2,11,2))
time=list(range(1,6))
# d = 2*t

plt.title("Distance Vs Time")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(time,distance,'yo')
plt.grid()
plt.show()
```

```python
# V = a * t
t= list(range(1,11))
a = list(range(1,len(t)+1))
v = [2*i for i in t]
plt.plot(t,v)
plt.grid()
plt.show()
```

```python
n = list(range(1,10))
sqn = [i**3 for i in n]
plt.title("Linear & Non-Linear Graph")
plt.plot(n,sqn)
plt.grid()
```

```python
# Voltage - Time
# Voltage = 1 + sin(2*pi*t)
import numpy as np
#t = [i/10 for i in range(1,21)]
t=np.arange(0.1,2.0,0.1)
v = [1+np.sin(2*np.pi*i) for i in t]
plt.title("Time Vs Voltage Graph")
plt.plot(t,v)
plt.grid()
plt.xlabel("Time")
plt.ylabel("Voltage")
plt.savefig("./Images/voltageTime.jpg")
```

```python
# Plot Income of Arkansas from 2005 to 2013
import pandas as pd
df = pd.read_csv("./DataFiles/Income.csv")
years = df.columns[2:]

incomeArkansas = df.values[3,2:]
plt.title=("Income of Arkansas")
plt.xlabel("Year")
plt.ylabel("Income")
plt.plot(years,incomeArkansas)
plt.show()
df
```

```python
# Income of All States from  2005 to 2013
incomeallstates = []
for i in df.columns[2:]:
    incomeallstates.append(df[i].mean())
plt.plot(years,incomeallstates)
#plt.title("All States Income")
plt.xlabel('Year')
plt.ylabel("Income")
plt.show()
```
