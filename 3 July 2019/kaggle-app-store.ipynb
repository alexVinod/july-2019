```python
# List of all unique Prime_Genres(categories) in the data sets.
def categoriesUnique():
    categories= appstore["prime_genre"].unique()
    return [i for i in categories]
for i in categoriesUnique():
    print(i,end="| ")
#appstore.lx[(appstore["prime_genre"])]
#appstore.loc[appstore["prime_genre"]=="Games"]
```

```python
# Category with highest number of apps
def highestNumberOfApps():
    highestApps={}
    for i in appstore["prime_genre"]:
        if i not in highestApps: highestApps[i]=1
        else:highestApps[i]+=1
    # max(highestApps, key=highestApps.get)
    return max(highestApps, key=highestApps.get),highestApps[max(highestApps, key=highestApps.get)]
    #print(max(highestApps, key=highestApps.get),highestApps[max(highestApps, key=highestApps.get)])
highestNumberOfApps()

# Category with lowest number of apps
def lowestNumberOfApps():
    lowesttApps={}
    for i in appstore["prime_genre"]:
        if i not in lowesttApps: lowesttApps[i]=1
        else:lowesttApps[i]+=1
    return (min(lowesttApps,key=lowesttApps.get),lowesttApps[min(lowesttApps,key=lowesttApps.get)])

#lowestNumberOfApps()
print(highestNumberOfApps(),lowestNumberOfApps())
```

```python
# Category with highest user rating
def categoryHighestUserRating():
    categoryHighestUserRating={}
    #maxUserRating=max(appstore['user_rating'])
    maxUserRating=appstore.loc[(appstore['user_rating']==max(appstore['user_rating']))]
    for i in maxUserRating["prime_genre"]:
        if i not in categoryHighestUserRating:
            categoryHighestUserRating[i]=1
        else:
            categoryHighestUserRating[i]+=1
    highUserRating = max(categoryHighestUserRating,key=categoryHighestUserRating.get)
    highUserRatingNumber = categoryHighestUserRating[max(categoryHighestUserRating,key=categoryHighestUserRating.get)]
    print("'{}' with highest user rating of '{}'".format(highUserRating,highUserRatingNumber))
    #return appstore["prime_genre"].value_counts().idmax() + str(appstore['prime_genre'].value_counts().max())
    
    #print(maxUserRating["prime_genre"].unique())
    #print("User Highest Rating : ",maxUserRating)
categoryHighestUserRating()
```

```python
# App with highest downloads
def appHighestDownloads():
    highestAppDownloads={}
    for i in range(len(appstore['track_name'])):
        highestAppDownloads[appstore['track_name'][i]]=appstore['rating_count_tot'][i]
    maxHighestDownloads=max(highestAppDownloads,key=highestAppDownloads.get)
    maxHighestDownloadsCount=highestAppDownloads[max(highestAppDownloads,key=highestAppDownloads.get)]
    print(maxHighestDownloads,maxHighestDownloadsCount)        
appHighestDownloads()
```

```python
# Average user rating for free apps
def avgUserRatingFreeApps():
    avgFreePrice=appstore.loc[(appstore['price']==0)]
    sumRating=sum(avgFreePrice["user_rating"])
    avgFreeApps=sumRating/len(avgFreePrice["user_rating"])
    print("Average user rating for free apps : '{}'".format(avgFreeApps))
avgUserRatingFreeApps()
```

```python
# Average user rating for paid apps
def avgUserRatingPaidApps():
    avgPaidPrice = appstore.loc[(appstore['price']!=0)]
    sumRate = sum(avgPaidPrice["user_rating"])
    avgPaidApps=sumRate/len(avgPaidPrice['user_rating'])
    print("Average user rating for paid apps : '{}'".format(avgPaidApps))
avgUserRatingPaidApps()
```

```python
# Most frequent Price Point > 0
def mostFrequentPricePointGreaterThanZero():
    greatThanZero=appstore.loc[(appstore["price"]>0)]
    mostFrequency=max(greatThanZero["price"].value_counts())
    print("Most frequent Price Point > 0 is : '{}'".format(mostFrequency))
mostFrequentPricePointGreaterThanZero()

```

```python
# Compare average user rating for paid vs free gaming apps
def compareAvgUserRatingPaidVsFreeGamingApps():
    #appstore.groupby(['prime_genre']).mean()
    freeGamingApps = appstore.loc[(appstore["price"]==0) & (appstore["prime_genre"]=="Games")]
    freeAvgRating = sum(freeGamingApps["user_rating"])/len(freeGamingApps["user_rating"])
    
    paidGamingApps = appstore.loc[(appstore["price"]>0) & (appstore["prime_genre"]=="Games")]
    paidAvgRating = sum(paidGamingApps["user_rating"])/len(paidGamingApps["user_rating"])
    print("Paid :{} \nFree : {}".format(paidAvgRating,freeAvgRating))
    #print(abs(paidAvgRating-freeAvgRating))
    
compareAvgUserRatingPaidVsFreeGamingApps()

#freeGamingApps = appstore.loc[(appstore["price"]==0) & (appstore["prime_genre"]=="Games")]
#freeGamingApps
```
